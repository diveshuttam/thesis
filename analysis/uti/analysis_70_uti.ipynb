# Analysis 70 UTI

```python
import csv
from pprint import pprint
import os
import numpy as np
```

```python
folder = "./output/processed/uti_w_tratio_summary_results"
error_file_list = os.listdir(folder)
# pprint(error_file_list)
momon_files = list(filter(lambda x: "momon" in x, error_file_list))
cemon_files = list(filter(lambda x: "cemon" in x, error_file_list))
```

```python
def convert(m1_):
    m1 = [m1_[1]]+list(map(lambda x:round(float(x),3),[m1_[18],m1_[20],m1_[22],m1_[24],m1_[26],m1_[4],m1_[6],m1_[11],m1_[13]]))
    m1.extend([m1[-4]*m1[-2],m1[-3]*m1[-1]])
    m1.append(round(m1[-1]/m1[-2],3))
    return m1
```

```python
momon_files_data = []
for f in momon_files:
    data = list(sorted(map(convert,csv.reader(open(f"{folder}/{f}")))))
    momon_files_data.append(data)
```

```python
cemon_files_data = []
for f in cemon_files:
    data = list(sorted(map(convert,csv.reader(open(f"{folder}/{f}")))))
    cemon_files_data.append(data)
```

```python
print(len(momon_files_data))
print(len(cemon_files_data))
print(len(momon_files_data[0]))
print(len(cemon_files_data[0]))
print(cemon_files_data[0][0])
print(momon_files_data[0][0])
```

# CeMon

```python
ans = [] 
for files_row in zip(*cemon_files_data):
    # print(len(files_row))
    s = set(map(lambda x:tuple(x[1:6]),files_row))
    assert(len(s)==1)
    # print(list(s))
    mi = min(map(lambda x:x[-1],files_row))
    ma = max(map(lambda x:x[-1],files_row))
    avg = np.average(list(map(lambda x:x[-1],files_row)))
    # print(mi)
    # print(avg)
    ans.append([list(s)[0], mi, ma, avg])
```

## Avg based cemon

```python
pprint(list(sorted(ans, key=lambda x:-x[-1]))[:60])
# tmin, tmax, param, ti, td, min, max, avg
```
