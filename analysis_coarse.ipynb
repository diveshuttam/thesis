# Analysis Coarse

```python
import csv
from pprint import pprint
```

```python
momon1 = list(sorted(list(csv.reader(open("./results/momon1.tsv"), delimiter='\t'))))
cemon1 = list(sorted(list(csv.reader(open("./results/cemon1.tsv"), delimiter='\t'))))
momon3 = list(sorted(list(csv.reader(open("./results/momon3.tsv"), delimiter='\t'))))
cemon3 = list(sorted(list(csv.reader(open("./results/cemon3.tsv"), delimiter='\t'))))
momon4 = list(sorted(list(csv.reader(open("./results/momon4.tsv"), delimiter='\t'))))
cemon4 = list(sorted(list(csv.reader(open("./results/cemon4.tsv"), delimiter='\t'))))
```

```python
print(len(momon1))
print(len(cemon1))
print(len(momon3))
print(len(cemon3))
print(len(momon4))
print(len(cemon4))
```

# CeMon

```python
ans=[]
for m1,c1,m3,c3,m4,c4 in zip(momon1,cemon1,momon3,cemon3,momon4,cemon4):
    # par = set(list(map(tuple,[m1[1:6],c1[1:6],m3[1:6],c3[1:6],m4[1:6],c4[1:6]])))
    # print(par, len(par))
    # assert(len(par)==1)
    m1=c1
    m3=c3
    m4=c4
    mi = min(map(float,[m1[-1],c1[-1],m3[-1],c3[-1],m4[-1],c4[-1]]))
    avg = sum(map(float,[m1[-1],c1[-1],m3[-1],c3[-1],m4[-1],c4[-1]]))/6
    ans.append([m1[1:6], mi, avg])
```

## Avg based cemon

```python
pprint(list(sorted(ans, key=lambda x:-x[-1]))[:30])
```

## Min based cemon

```python
pprint(list(sorted(ans, key= lambda x:-x[-2]))[:60])
```
